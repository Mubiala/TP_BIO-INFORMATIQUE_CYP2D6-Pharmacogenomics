```python
#Construction de la Base d'interprétation de clinique

interpretation = {
    "No function": {
        "Phenotype": "Poor Metabolizer",
        "Medication": "Codeine, Tramadol",
        "Recommendation": "Avoid use; choose an alternative drug."
    },

    "Decreased function": {
        "Phenotype": "Intermediate Metabolizer",
        "Medication": "Tamoxifen, Antidepressants",
        "Recommendation": "Dose adjustment may be required."
    },

    "Normal function": {
        "Phenotype": "Normal Metabolizer",
        "Medication": "Most CYP2D6 substrates",
        "Recommendation": "Use the standard recommended dose."
    },

    "Increased function": {
        "Phenotype": "Ultrarapid Metabolizer",
        "Medication": "Codeine, Tramadol",
        "Recommendation": "Avoid use due to toxicity risk."
    },

    "Unknown function": {
        "Phenotype": "Unknown",
        "Medication": "Unknown",
        "Recommendation": "Further genetic investigation required."
    },

    "Uncertain function": {
        "Phenotype": "Unknown",
        "Medication": "Unknown",
        "Recommendation": "Clinical significance not yet established."
    }
}
```

```python
#Extraction de noms des médicaments

import re
from collections import Counter

# Récupération de toutes les lignes non nulles
texts = cyp2d6["Findings"].dropna().astype(str)

# Extraction des médicaments
medicaments = []

for texte in texts:
    # On enlève les PMID
    texte = re.sub(r"\d+:", "", texte)

    # Séparer les éléments
    parties = texte.split(";")

    for p in parties:
        p = p.strip()

        # Supprimer les mentions (in vivo), (in vitro), etc.
        p = re.sub(r"\(.*?\)", "", p)

        # Séparer les médicaments lorsqu'il y en a plusieurs
        for med in p.split(","):
            med = med.strip()

            if len(med) > 2:
                medicaments.append(med)

# Comptage
compteur = Counter(medicaments)

# Les 20 plus fréquents
top20 = compteur.most_common(20)

top20
```

```python
#filtrage de ces résultats pour ne conserver que les véritables médicaments.

# Liste des médicaments d'intérêt
medicaments_valides = [
    "dextromethorphan",
    "bufuralol",
    "debrisoquine",
    "sparteine",
    "tamoxifen",
    "N-desmethyltamoxifen",
    "atomoxetine",
    "risperidone",
    "venlafaxine",
    "amitriptyline",
    "citalopram",
    "codeine",
    "metoprolol",
    "primaquine"
]

# Filtrer les résultats
top_medicaments = {
    med: compteur[med]
    for med in medicaments_valides
    if compteur[med] > 0
}

top_medicaments
```

```python
#Graphique

plt.figure(figsize=(10,6))

plt.bar(
    top_medicaments.keys(),
    top_medicaments.values()
)

plt.title("Médicaments associés aux variants du gène CYP2D6")

plt.xlabel("Médicament")

plt.ylabel("Nombre d'occurrences")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```
